計算每個欄位的異常值

```python
import pandas as pd

# 定義一個函數來計算每個欄位的異常值，並返回上下邊界值
def detect_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)]
    return outliers[column], lower_bound, upper_bound

# 初始化結果存儲結構
outliers_summary = []

# 遍歷 data1 到 data17
for i in range(1, 18):
    data_name = f"data{i}"
    
    # 假設你的 data 都是 pandas dataframe 形式存放
    dataframe = globals()[data_name]  # 使用 globals() 來動態加載變量
    
    # 對每個欄位計算 outliers
    for column in ['WindSpeed(m/s)', 'Pressure(hpa)', 'Temperature(°C)', 'Humidity(%)', 'Sunlight(Lux)', 'Power(mW)']:
        outliers, lower_bound, upper_bound = detect_outliers_iqr(dataframe, column)
        outliers_summary.append({
            'Dataset': data_name,
            'Column': column,
            'Outlier Count': len(outliers),
            'Lower Bound': lower_bound,
            'Upper Bound': upper_bound
        })

# 將結果轉換為 DataFrame 並顯示
outliers_summary_df = pd.DataFrame(outliers_summary)
print(outliers_summary_df)
```

```python
#對data1-data17中的'WindSpeed(m/s)', 'Pressure(hpa)','Temperature(°C)', 'Humidity(%)', 'Sunlight(Lux)', 'Power(mW)'畫直方圖，想把同個變項的直方圖放在一起比較
import pandas as pd
import matplotlib.pyplot as plt

# 定義要繪製直方圖的數據集
data_list = [data1, data2, data3, data4, data5, data6, data7, data8, data9, data10, data11, data12, data13, data14, data15, data16, data17]

# 定義一個函數來繪製某個變項的直方圖
def plot_histograms(variable_name):
    plt.figure(figsize=(10, 6))

    # 統一範圍
    min_value = min(data[variable_name].min() for data in data_list)
    max_value = max(data[variable_name].max() for data in data_list)
    data_range = (min_value, max_value)

    # 依次繪製每個數據集的該變項的直方圖
    for i, data in enumerate(data_list):
        plt.hist(data[variable_name], bins=20, alpha=0.5, label=f'data{i+1}', histtype='stepfilled', range=data_range)
    
    #不要grid
    plt.grid(False)
    plt.title(f'Histogram of {variable_name}')
    plt.xlabel(variable_name)
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.show()


# 要繪製的變項列表
variables_to_plot = ['WindSpeed(m/s)', 'Pressure(hpa)', 'Temperature(°C)', 'Humidity(%)', 'Sunlight(Lux)', 'Power(mW)']

# 為每個變項繪製直方圖
for variable in variables_to_plot:
    plot_histograms(variable)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 定義要繪製折線圖的數據集
data_list = [data1, data2, data3, data4, data5, data6, data7, data8, data9, data10, data11, data12, data13, data14, data15, data16, data17]

# 定義一個函數來繪製某個變項的折線圖
def plot_line_graphs(variable_name):
    plt.figure(figsize=(10, 6))

    # 統一範圍
    min_value = min(data[variable_name].min() for data in data_list)
    max_value = max(data[variable_name].max() for data in data_list)
    bins = np.linspace(min_value, max_value, 20)

    # 依次繪製每個數據集的該變項的折線圖
    for i, data in enumerate(data_list):
        counts, _ = np.histogram(data[variable_name], bins=bins)
        bin_centers = (bins[:-1] + bins[1:]) / 2  # 計算每個bin的中心點
        plt.plot(bin_centers, counts, marker='o', label=f'data{i+1}', alpha=0.7)

    #不要grid
    plt.grid(False)
    plt.title(f'Line Plot of {variable_name}')
    plt.xlabel(variable_name)
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.show()

# 要繪製的變項列表
variables_to_plot = ['WindSpeed(m/s)', 'Pressure(hpa)', 'Temperature(°C)', 'Humidity(%)', 'Sunlight(Lux)', 'Power(mW)']

# 為每個變項繪製折線圖
for variable in variables_to_plot:
    plot_line_graphs(variable)
```
