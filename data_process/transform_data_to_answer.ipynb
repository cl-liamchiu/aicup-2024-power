```python
predictions_df = pd.read_csv('../result/result_72250.csv')

# 只選擇與 LocationCode 相關的欄位
location_columns = [col for col in predictions_df.columns if 'LocationCode_' in col]

# 使用 idxmax 找出每一行最大值的索引
location_code = predictions_df[location_columns].idxmax(axis=1)

# 去除 'LocationCode_' 前綴
location_code = location_code.str.replace('LocationCode_', '')

print(location_code)

# 將結果加入回原始 DataFrame
predictions_df['LocationCode'] = location_code
#把result中LocationCode_1 、LocationCode_2、...、LocationCode_17
predictions_df = predictions_df.drop(['LocationCode_1', 'LocationCode_2', 'LocationCode_3', 'LocationCode_4', 'LocationCode_5', 'LocationCode_6', 'LocationCode_7', 'LocationCode_8', 'LocationCode_9', 'LocationCode_10', 'LocationCode_11', 'LocationCode_12', 'LocationCode_13', 'LocationCode_14', 'LocationCode_15', 'LocationCode_16', 'LocationCode_17'], axis=1)
print(predictions_df)
```

```python
data_after_72250_test = pd.read_csv('../data/preprocess_data/oneday_weather_6timestep/data_after_72250_test_6power.csv') 
result_y_pred = data_after_72250_test.copy()

#將filtered_result_111的index重設
result_y_pred.reset_index(drop=True, inplace=True)

#把預測結果加入result_y_pred
result_y_pred['Power_1'] = predictions_df['Power_1']
result_y_pred['Power_2'] = predictions_df['Power_2']
result_y_pred['Power_3'] = predictions_df['Power_3']
result_y_pred['Power_4'] = predictions_df['Power_4']
result_y_pred['Power_5'] = predictions_df['Power_5']
result_y_pred['Power_6'] = predictions_df['Power_6']

result_y_pred
```

```python
# 定義完整的時間範圍
time_range = list(range(9, 17))

# 補齊缺失時間的函數
def fill_missing_times(group):
    # 建立完整的時間 DataFrame
    complete_times = pd.DataFrame({
        "Time": time_range
    })
    # 合併原始資料並補充缺失行
    merged = pd.merge(complete_times, group, on="Time", how="left")
    # 填補 LocationCode 和 Date 的值
    merged["LocationCode"] = merged["LocationCode"].fillna(group["LocationCode"].iloc[0])
    merged["Date"] = merged["Date"].fillna(group["Date"].iloc[0])
    return merged

# 對每組進行補齊操作
result = (
    result_y_pred.groupby(["LocationCode", "Date"])
    .apply(fill_missing_times)
    .reset_index(drop=True)
)

# 結果展示
print(result)
```

```python
#把result中第10欄位以後刪除
result = result.drop(result.columns[9:], axis=1)
result
```

```python
# 定義接續函數
def concatenate_power_to_columns(group):
    # 將 Power_1 ~ Power_6 按時間順序展平成單一列表
    power_columns = [f"Power_{i}" for i in range(1, 7)]
    concatenated = group[power_columns].values.flatten()
    # 將數據展開為多列
    result = {f"Power_{i+1}": val for i, val in enumerate(concatenated)}
    return pd.Series(result)

# 分組並應用接續函數
result_1 = (
    result
    .groupby(["LocationCode", "Date"])
    .apply(concatenate_power_to_columns)
    .reset_index()
)

# 結果展示
print(result_1)
```

```python
result = result_1.copy()
```

```python
data_sub = pd.read_csv('../result/upload(no answer).csv')

data_sub['序號'] = data_sub['序號'].astype(str)

#把data_sub創建一欄LocationCode，值為'序號'欄位末兩碼
data_sub['LocationCode'] = data_sub['序號'].str[-2:]

#把data_sub創建一欄Date，值為'序號'欄位前8碼
data_sub['Date'] = data_sub['序號'].str[:8]

#把data_sub創建一欄Time，值為'序號'欄位第9-12碼
data_sub['Time'] = data_sub['序號'].str[8:12]

#把data_sub第三四五欄位，存成data_sub_1
data_sub_1 = data_sub.iloc[:, 2:5]

data_sub_1['Value'] = 1  # 用 1 作為示例值

# 轉置資料
result_1 = data_sub_1.pivot(index=['LocationCode', 'Date'], columns='Time', values='Value')

# 重設欄位索引
result_1.reset_index(inplace=True)

#把result_1中的Date欄位改成2024-01-06，現在是20240106
result_1['Date'] = pd.to_datetime(result_1['Date'], format='%Y%m%d')
result_1['Date'] = result_1['Date'].dt.strftime('%Y-%m-%d')
result['Date'] = pd.to_datetime(result['Date'], format='%Y-%m-%d')
result['Date'] = result['Date'].dt.strftime('%Y-%m-%d')

result_1['LocationCode'] = result_1['LocationCode'].astype(int)
result['LocationCode'] = result['LocationCode'].astype(int)

#把result_1依照LocationCode,Date和result中的LocationCode,Date_Day2合併
result_final = pd.merge(result_1, result, left_on=['LocationCode', 'Date'], right_on=['LocationCode', 'Date'], how='left')
result_final

```

```python

# 刪除result_final中的第3-50欄位
result_final = result_final.drop(columns=result_final.columns[2:50])
result_final

#把result_final中第3欄到第50欄位名稱依序改為0900、0910、0920、...、1650
result_final.columns = ['LocationCode', 'Date', '0900', '0910', '0920', '0930', '0940', '0950', '1000', '1010', '1020', '1030', '1040', '1050', '1100', '1110', '1120', '1130', '1140', '1150', '1200', '1210', '1220', '1230', '1240', '1250', '1300', '1310', '1320', '1330', '1340', '1350', '1400', '1410', '1420', '1430', '1440', '1450', '1500', '1510', '1520', '1530', '1540', '1550', '1600', '1610', '1620', '1630', '1640', '1650']
final_result = pd.DataFrame(result_final)


# 將數據轉換為長格式（melt），使每個時間欄位展平為單獨的行
final_result_long = final_result.melt(
    id_vars=["LocationCode", "Date"],  # 保留這些欄位不變
    var_name="Time",                  # 時間欄位名稱
    value_name="Power"                # 值的名稱
)


#把data_sub的Date欄位改成2024-01-06，現在是20240106
data_sub['Date'] = pd.to_datetime(data_sub['Date'], format='%Y%m%d')

#把data_sub中的LocationCode 轉成float32
data_sub['LocationCode'] = data_sub['LocationCode'].astype(float)
data_sub['Time'] = data_sub['Time'].astype(float)
data_sub.dtypes
# 將 Date 欄位轉換為 datetime 格式，確保兩者一致
final_result_long['Date'] = pd.to_datetime(final_result_long['Date'])
data_sub['Date'] = pd.to_datetime(data_sub['Date'])

# 確保 Time 欄位格式一致，將數字轉為字符串並補全格式
final_result_long['Time'] = final_result_long['Time'].astype(str)
data_sub['Time'] = data_sub['Time'].astype(int).astype(str).str.zfill(4)

# 合併 final_result_long 和 data_sub，根據 LocationCode, Date, Time 匹配
data_sub = data_sub.merge(
    final_result_long,
    on=['LocationCode', 'Date', 'Time'],  # 依據這些欄位進行匹配
    how='left'  # 保留 data_sub 的所有行，若無匹配則填充 NaN
)

# 查看結果
print(data_sub)
```
